```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# in our data there is only country code is avalibel so we are creating a new column "Country" in our data. And for this we are loading a another data set 
country_code=pd.read_excel("Country-Code.xlsx")
```

```python
country_code
```

```python
# now merge manin data set and country_code data set and create a new column in main data set "Country", in this column we are inputing correct country name according to the country code
data=pd.merge(data, country_code,on='Country Code',how='left')
```

```python
data.columns
```

### EDA 

#### 1. Zomato usage by Country

```python
country_count=data.Country.value_counts()
country_count
```

```python
highest_country=country_count.idxmax()
highest_country
```

```python
plt.figure(figsize=(25,25))
labels = country_count.index[:6]
plt.pie(country_count[:6], labels=labels,autopct='%1.2f%%', startangle=140, pctdistance=.9,textprops={'rotation': -45})
plt.legend(labels, loc="best")
plt.title(" Top 6 Country in distribution",fontsize=30)
plt.show()
```

* From this we can observe that zomato has maximum records of transaction from are from india after that USA and then United Kingdom

#### 2. Ratings by customers

```python
data.columns
```

```python
# creating a new data set containg 'Aggregate rating','Rating color' and 'Rating text' and also make a count on ratings 
rating=data.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
rating
```

* 0 rating color with White labeled as Not rated
* 0 - 2.4 rating color with Red labeled as Poor
* 2.5 - 3.4 rating color with Orange labeled as Average
* 3.5 - 3.9 rating with color with Yellow labeled as Good
* 4 - 4.4 rating with color Green labeled as Very Good
* 4.5 - 5 rating with color Dark Green as Excellent

```python
plt.figure(figsize=(15,10))
colors = sns.color_palette("husl", len(rating['Aggregate rating']))
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Aggregate rating',data=rating,palette=colors)
plt.title("Count Of Rating")
plt.show()
```

* Most of the peoples are not rated 
* in rating , the most people rated in betweeen 3.0 - 3.8 


```python
sum_of_rate=rating.groupby(['Rating color', 'Rating text'])['Rating Count'].sum().reset_index()
sum_of_rate
```

```python
plt.figure(figsize=(10,6))
color=['#006400', '#008000', '#FFA500', '#FF0000', '#0000FF', '#FFFF00']
sns.barplot(x='Rating color',y='Rating Count',hue='Rating text',data=sum_of_rate,palette=color,legend=True)
plt.title("Count Of Rating Color")
plt.show()
```

* from this we can understand that most transaction rated as Average and least rated as Poor. And the main thing is count on "Not rated" also a big number
